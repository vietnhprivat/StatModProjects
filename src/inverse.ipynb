```python
import sympy as sp

# 1. Initialize LaTeX printing (This is the key step)
sp.init_printing()

# 2. Define symbols
y, lam, z = sp.symbols('y lambda z', positive=True)

# 3. Define the expression (Equation implied = 0)
expr = (1/lam) * sp.log((y**lam) / (1 - y**lam)) - z

# 4. Solve for y
solution = sp.solve(expr, y)

# 5. Display the result
# solution is a list, so we display the first element
display(solution[0])
```
